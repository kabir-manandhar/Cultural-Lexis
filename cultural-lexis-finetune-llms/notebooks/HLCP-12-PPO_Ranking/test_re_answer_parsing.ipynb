````python
en_data = "asdfsdfs\nsdfsdf\nsdfs\n Final Ranking:\nRank 1: [Associated Word]\nRank 2: [Associated Word]\n...\nRank K: [Associated Word]\n"

zh_data = "\n\n```\n最终排名:\n排名1: [关联词]\n排名2: [关联词]\n...\n排名120: [关联词]\n```\n"
````

````python
import re
````

````python
answer_line = en_data.lower().rfind("final ranking")
answer = en_data[answer_line + len("Final Ranking"):]
found_lst = re.findall(r"(\d+)\s*[:：]\s*(.*)", answer)
parsed_dict= dict()
for rank, word in found_lst:
    if int(rank) not in parsed_dict:
        parsed_dict[int(rank)] = word.strip()
# sort the dict by key
parsed_dict = sorted(parsed_dict.items(), key=lambda x: x[0])
parsed_dict = dict(parsed_dict)
print(answer)
print(parsed_dict)
````
